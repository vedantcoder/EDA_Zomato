```python
# Finding missing values
df.isnull().sum()
```

```python
# Find feature with missing values
[features for features in df.columns if df[features].isnull().sum()>0]
```

```python
#Heatmap
mat.rcParams['figure.figsize'] = (12,6)
sns.heatmap(df.isnull(),yticklabels=False, cbar=False, cmap='viridis')
```

```python
# Loading and viewing the Country code sheet
df_country=pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
# Merge the 2 sheets and focus on the original one
final_df= pd.merge(df,df_country, on='Country Code', how='left')
final_df.head()
```

```python
# Another way to check datatypes
final_df.dtypes
```

```python
final_df.columns
```

```python
Country_name = final_df.Country.value_counts().index
Country_name
```

```python
Country_value = final_df.Country.value_counts().values
Country_value
```

```python
final_df.Country.value_counts()
```

```python
# Pie chart to show division by country
# plt.pie(Country_value, labels=Country_name)

## Since it is a lot of crowding towards the end (very small values for a lot of countries we will look at top 3)
plt.pie(Country_value[:3], labels=Country_name[:3], autopct='%1.2f%%')
```

--observations--

```python
# To see ratings and the corresponding color & comments
ratings= final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Count'})
ratings
```

## Conclusions
1 Rating >= 4.5 = Excellent

2 Rating >= 4.0 = Very Good

3 Rating >= 3.5 = Good

4 Rating >= 2.5 = Average

5 Rating >= 1.8 = Poor

6 A lot of people have not rated the service!

```python
ratings.head()
```

```python
mat.rcParams['figure.figsize'] = (12,6)
sns.barplot(x="Aggregate rating", y="Count", data=ratings, hue='Rating color', palette=['black', 'red', 'orange', 'yellow', 'green', 'green'])
```

--observations--

```python
## Count plot
sns.countplot(x="Rating color", data=ratings, hue='Rating color', palette=['black', 'red', 'orange', 'yellow', 'green', 'green'])
```
